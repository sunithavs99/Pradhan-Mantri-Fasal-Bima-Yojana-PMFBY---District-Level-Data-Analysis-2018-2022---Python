#### 2.2.1 Data Type Correction

- Categorical data types are memory-efficient representations of repeated string values, whereas object data types store raw text and consume more memory. So the object data type column can be converted to categorical data type wherever necessary.
- The ***'year, season, scheme, state_name, state_code, district_name, district_code'*** columns can be convereted to categorical columns based on the nature of data holding by the columns. Converting these columns into categorical data types improves memory efficiency and enables more effective grouping and analysis during exploratory data analysis.

```python
# Convert year, season, scheme, state_name, state_code, district_name, district_code columns to categorical data type
categorical_cols = [
    'year', 'season', 'scheme',
    'state_name', 'state_code',
    'district_name', 'district_code'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

- Certain columns namely ***'farmer_count, loanee, non_loanee, iu_count'*** representing counts are stored as floating-point values, which is not appropriate for numerical count data. These columns are converted to integer data types. Correcting data types improves data accuracy, reduces memory usage, and ensures reliable analysis.

```python
# Convert farmer_count, non_loanee, iu_count columns to integer data type
count_cols = ['farmer_count', 'non_loanee', 'iu_count']

for col in count_cols:
    df[col] = df[col].astype(int)
```

- **Handling Missing Value in Loanee Column Using Mean Imputation:** The ***'loanee'*** column contains one missing value. Inorder to correct the data type in this column, the missing value also need to handle. Since the column type is of a count and planning to convert into integer data type. Since only a single missing value is present in this column, mean imputation is applied to replace the missing entry. This approach preserves the record while minimizing the impact on overall data distribution. Mean imputation is appropriate in this context as it ensures continuity of numerical analysis without introducing significant bias.

```python
# Calculate mean of loanee column
loanee_mean = df['loanee'].mean()

# Replace missing value with mean and convert to the integer data type
df['loanee'] = df['loanee'].fillna(loanee_mean).round().astype(int)
```

- Final updated data types:

```python
# Verify updated data types
df.dtypes
```

- **Observation:** Relevant object-type columns have been successfully converted to categorical data types, and count-based columns have been corrected to integer types. These transformations improve memory efficiency, ensure numerical correctness, and enhance the reliability of further exploratory and statistical analysis.

#### 2.2.2 Handling Missing Values

- The ***'district_code'*** column is stored as a categorical data type. In Pandas, categorical columns do not allow insertion of new values unless they are explicitly defined as categories. Therefore, the category "Unknown" is first added to the existing category list before replacing missing values. This ensures proper handling of missing district identifiers while maintaining categorical integrity.
- Column ***'district_code'*** contains 739 missing values and this column represents an administrative identifier rather than a numerical measure. Missing values in this column are handled by assigning the label "Unknown" to clearly indicate unavailable district information. This approach avoids introducing misleading numeric values and ensures accurate grouping and categorization during analysis.

```python
# Add 'Unknown' as a valid category
df['district_code'] = df['district_code'].cat.add_categories(['Unknown'])

# Fill missing values with 'Unknown'
df['district_code'] = df['district_code'].fillna('Unknown')
```

- **Standardizing District Code Format:** The district_code column originally contained numeric values with missing entries, causing Pandas to store the column as floating-point data. As a result, district codes appeared with decimal points (e.g., 101.0). Since district codes are identifiers and not numerical measures, the column is converted to string format to remove decimal notation and then cast back to a categorical data type. This ensures correct representation and consistent categorization of district identifiers.

```python
# Convert district_code to string, remove .0, then convert to category
df['district_code'] = (
    df['district_code']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
    .astype('category')
)
```

- The ***'loanee'*** column contains one missing value and it was replaced by mean during data type correction in step 2.2.1

- **Handling Missing Values in Percentage Columns:** The columns ***'male, female, transgender, sc, st, obc, gen, marginal, small, and other'*** represent percentage distributions of beneficiaries across demographic and farmer categories. These values are measured as proportions rather than absolute counts. There are ***70*** missing counts in each of these columns.

    Missing values in these columns indicate that percentage information was not reported for the corresponding records. To maintain numerical consistency and avoid misinterpretation, missing values are replaced with 0.0, indicating zero percentage contribution and all values are rounded to two decimal places for consistency and improved readability for the respective category. The float data type is retained to preserve precision for percentage-based analysis.

```python
# Demographic percentage columns
percentage_cols = [
    'male', 'female', 'transgender',
    'sc', 'st', 'obc', 'gen',
    'marginal', 'small', 'other'
]

# Fill missing values, retain float, and round to 2 decimal places
df[percentage_cols] = df[percentage_cols].fillna(0.0).round(2)
```

- **Rounding All Float Columns to Two Decimal Places:** To maintain numerical consistency and improve readability, all columns with floating point data types are rounded to two decimal places. This standardization ensures uniform representation of decimal values across the dataset while preserving analytical accuracy for percentage and value based computations.

```python
# Identify all float columns in the dataset
float_cols = df.select_dtypes(include=['float']).columns

# Round all float columns to 2 decimal places
df[float_cols] = df[float_cols].round(2)
```

- **Missing Values Handling – Final Status:** All missing values in the dataset have been systematically identified and handled using appropriate strategies based on the nature of each column. Identifier columns were assigned meaningful labels, percentage-based columns were filled with zero values while retaining float precision, and numerical columns were filled with mean values to ensure analytical accuracy. The dataset is now free of missing values and ready for exploratory data analysis.

```python
# Display remaining missing values in the dataset
df.isna().sum()
```

- **Observation:** The missing value check confirms that all columns in the dataset contain zero missing values. This indicates successful completion of the missing value handling process, ensuring data completeness and reliability for subsequent analysis.

### 2.3 Creating Derived Columns

Derived columns are created to enhance analytical depth by transforming or combining existing variables. These additional features help in summarizing information, identifying trends, and improving interpretability during exploratory data analysis.

#### 2.3.1 Derived Column 1 - Total Farmers Covered

- **Total Farmers Covered:** The total number of farmers covered under the scheme is calculated by combining loanee and non-loanee farmers. This derived metric provides a consolidated view of farmer participation across districts and years.

```python
# Create total farmers covered column
df['total_farmers'] = df['loanee'] + df['non_loanee']

# Display first few rows
df[['loanee', 'non_loanee', 'total_farmers']].head()
```

#### 2.3.2 Derived Column 2 - Gender Dominance Gap Percentage

- **Gender Dominance Gap Percentage:** This column captures the difference between male and female farmer participation percentages, indicating gender disparity across districts.

```python
# Create gender dominance gap percentage column
df['gender_gap_pct'] = (df['male'] - df['female']).round(2)

# Display first few rows
df[['male', 'female', 'gender_gap_pct']].head()
```

#### 2.3.3 Derived Column 3 - Small & Marginal Farmer Share Percentage

- **Small & Marginal Farmer Share Percentage:** This column combines marginal and small farmer percentages to assess scheme inclusiveness among vulnerable farmer groups.

```python
# Calculating Small & Marginal Farmer Share
df['small_marginal_share_pct'] = (df['marginal'] + df['small']).round(2)

# Display first few rows
df[['small_marginal_share_pct']].head()
```

#### 2.3.4 Derived Column 4 - Loanee Share Percentage

- **Loanee Share Percentage:** This column measures the proportion of loanee farmers among total insured farmers, helping assess dependency on institutional credit across districts.

```python
# Calculating Loanee share percentage
df['loanee_share_pct'] = ((df['loanee'] / df['total_farmers']) * 100).fillna(0).round(2)

# Display first few rows
df[['loanee_share_pct']].head()
```

#### 2.3.5 Derived Column 5 - Non-Loanee Share Percentage

- **Non-Loanee Share Percentage:** This derived metric indicates the participation of non-loanee farmers, useful for understanding scheme outreach beyond institutional credit.

```python
# Calculating Non-Loanee share percentage
df['non_loanee_share_pct'] = ((df['non_loanee'] / df['total_farmers']) * 100).fillna(0).round(2)

# Display first few rows
df[['non_loanee_share_pct']].head()
```

***info()*** function gives the overall columns info to verify the derived columns created as expected.

```python
# Dataset Information
df.info()
```

        - Gross premium has an almost perfect correlation with GoI share (0.99) and state share (0.99), reflecting the subsidy-driven structure of PMFBY where government contributions dominate premium financing.
    - *Gender Participation Dynamics*
        - Gender gap percentage is very strongly correlated with male participation (0.93) and strongly negatively correlated with female participation (-0.88), confirming that higher male dominance directly widens the gender gap.
        - Female participation shows weak or negligible correlation with financial indicators, suggesting that increased insurance spending does not automatically translate into improved female inclusion.
        - Transgender participation exhibits near-zero correlation with most variables, indicating minimal influence on overall scheme metrics due to low representation.
    - *Social Category Relationships*
        - OBC participation shows a moderate positive correlation with sum insured (0.18) and gross premium (0.15), implying relatively higher scheme penetration among OBC farmers.
        - ST participation has a negative correlation with financial variables such as sum insured and premium, pointing towards comparatively lower insurance coverage in tribal-dominated regions.
        - Strong negative correlations among SC, ST, OBC, and General categories reflect their mutually exclusive population shares, which is statistically expected.
    - *Landholding Size & Farmer Structure*
        - Small and marginal farmer share is strongly positively correlated (0.71) with small farmer proportion, indicating dominance of smallholders in PMFBY enrollment.
        - Other landholding category shows a strong negative correlation (-0.88) with small & marginal share, highlighting the inverse relationship between large farmers and smallholder participation.
        - Higher small and marginal farmer share is associated with lower gender gap, suggesting relatively better female participation among smallholder households.
    - *Credit (Loanee vs Non-Loanee) Structure*
        - Loanee share and non-loanee share are almost perfectly negatively correlated (-0.96), which is expected as they are complementary components.
        - Loanee share has a moderate positive correlation with male participation (0.31) and gender gap (0.30), indicating higher male dominance among credit-linked farmers.
        - Non-loanee participation shows a relatively higher association with female participation, suggesting voluntary enrollment may be more inclusive.

- **Overall Interpretation:**

  Overall, the correlation analysis indicates that financial and coverage-related variables under PMFBY move together, while demographic and social variables reveal structural inequalities in participation. The dominance of small and marginal farmers highlights the scheme’s pro-poor orientation, whereas persistent gender and regional disparities suggest the need for targeted policy interventions.

## **Step5-Insight Generation and Report**

### 5.1 Overall Findings
The PMFBY scheme has helped increase crop insurance coverage and provided financial support to farmers. However, participation is mainly driven by loanee farmers, and women farmers benefit less compared to men. Some districts perform well, while others need improvement.

### 5.2 Key Insights from the Analysis

#### Insight 1: Increase in Farmer Participation Over the Years

***Finding:*** The total number of farmers covered under PMFBY has increased consistently from 2018 to 2022.

***Interpretation:*** This indicates growing awareness of crop insurance and improved reach of the scheme among farmers.

***Strategic Action:*** Continue awareness campaigns and strengthen enrollment drives during peak agricultural seasons.

***Business Recommendation:*** The government and insurers should maintain current outreach strategies and expand them to low-performing districts.

---

#### Insight 2: Fluctuation in Sum Insured and Gross Premium

***Finding:*** The sum insured and gross premium collected show year-to-year variations, with a peak around 2020.

***Interpretation:*** This suggests that insurance coverage and risk exposure increased during certain years, possibly due to climatic uncertainty.

***Strategic Action:*** Regular monitoring of risk-prone years and adjustment of premium structures accordingly.

***Business Recommendation:*** Insurers should design flexible premium models and improve risk assessment using historical data.

---

#### Insight 3: Dominance of Loanee Farmers

***Finding:*** Loanee farmers contribute the highest share of insured farmers across all years.

***Interpretation:*** Enrollment is largely driven by institutional credit linkage, making insurance almost compulsory for loanee farmers.

***Strategic Action:*** Introduce incentives and simplified processes for non-loanee farmers.

***Business Recommendation:*** Promote voluntary insurance enrollment through subsidies or awareness programs for non-loanee farmers.

---

#### Insight 4: Low Participation of Non-Loanee Farmers

***Finding:*** Non-loanee farmer participation remains relatively low compared to loanee farmers.

***Interpretation:*** Independent farmers may lack awareness or perceive insurance as unnecessary.

***Strategic Action:*** Conduct village-level education programs highlighting the benefits of crop insurance.

***Business Recommendation:*** Develop customized insurance products targeted specifically at non-loanee farmers.

---

#### Insight 5: Gender Gap in Farmer Participation

***Finding:*** Male farmers consistently outnumber female farmers in insurance enrollment, resulting in a high gender gap.

***Interpretation:*** Women farmers face barriers such as lack of access to credit, information, and ownership records.

***Strategic Action:*** Promote women-focused enrollment drives and simplify documentation requirements.

***Business Recommendation:*** Introduce special schemes or incentives to encourage women farmers to enroll in PMFBY.

---

#### Insight 6: District-Level Variations in Coverage

***Finding:*** Significant variation exists among districts in terms of farmer coverage, sum insured, and premium collected.

***Interpretation:*** This indicates uneven implementation and varying levels of awareness across districts.

***Strategic Action:*** Identify low-performing districts and implement targeted interventions.

***Business Recommendation:*** Allocate more resources and manpower to districts with poor insurance penetration.

---

### 5.3. Final Conclusion

The analysis of the Pradhan Mantri Fasal Bima Yojana (PMFBY) data shows that the scheme has successfully increased farmer participation and insurance coverage over time. However, the findings also highlight key challenges such as heavy dependence on loanee farmers, low participation of non-loanee and women farmers, and uneven district-level performance.

To improve the effectiveness of the scheme, focused efforts are required to promote voluntary enrollment, reduce the gender gap, and strengthen outreach in low-performing districts. By implementing targeted strategies and inclusive policies, PMFBY can provide better financial protection and long-term stability to farmers across India.
